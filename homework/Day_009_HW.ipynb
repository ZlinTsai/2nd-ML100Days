```python
# 載入基礎套件
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

### 弱相關

```python
# 隨機生成兩組 1000 個介於 0~50 的數的整數 x, y, 看看相關矩陣如何
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)

# 呼叫 numpy 裡的相關矩陣函數 (corrcoef)
np.corrcoef(x, y)
```

```python
# 將分布畫出來看看吧
plt.scatter(x, y)
```

### 正相關

```python
# 隨機生成 1000 個介於 0~50 的數 x
x = np.random.randint(0, 50, 1000)

# 這次讓 y 與 x 正相關，再增加一些雜訊
y = x + np.random.normal(0, 10, 1000)

# 再次用 numpy 裡的函數來計算相關係數
np.corrcoef(x, y)
```

```python
# 再看看正相關的 x,y 分布
plt.scatter(x, y)
```

### 負相關

```python
x2 = np.random.randint(0, 50, 1000)
y2 = -x2 + np.random.normal(0, 10, 1000)

np.corrcoef(x2, y2)
```

```python
plt.scatter(x2, y2)
```

### check 

```python
def corr(x, y):
    n = len(x)
    mean_x = x.mean()
    mean_y = y.mean()
    temp_up = 0
    temp_botx = 0
    temp_boty = 0
    
    for i in np.arange(0, n, 1):
        temp_up  += (x[i] - mean_x)*(y[i] - mean_y)
        temp_botx += (x[i] - mean_x)**2
        temp_boty += (y[i] - mean_y)**2
        
    r = temp_up / np.sqrt(temp_botx*temp_boty)
    
    return r
```

```python
corr(x, y)
```

```python
corr(x2, y2)
```
